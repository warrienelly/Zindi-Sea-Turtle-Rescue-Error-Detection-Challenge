```python
# Function to get clean nd dirty_data that are different
def data_diff(clean, dirty):
    ne_stacked = (clean!= dirty).stack()
    changed = ne_stacked[ne_stacked]
    changed.index.names = ['id', 'col']
    difference_locations  = np.where(clean!= dirty)
    new_dirty = dirty.values[difference_locations]
    new_clean = clean.values[difference_locations]
    return pd.DataFrame({'dirt': new_dirty, 'clean': new_clean}, index=changed.index)
#     return pd.DataFrame(new_clean, columns=changed.index, index=['clean'])    

    
```

```python
# fuction to get data frame with more than one species
def plenty_species(df):
    # Get a dataframe of Tag_1 and species, grouped by Tag_1
    df_Tag_1_Species = df.groupby('Tag_1')['Species'].apply(lambda x: x.unique()).to_frame().reset_index()
    #check the unique values of each species
    #Locate values that contain more than one Species, 
    df_Tag_1_Species.Species = df_Tag_1_Species.Species.astype('str')
#     df_Tag_1_Species.Species.value_counts()
#     df_Tag_1_Species.loc[df_Tag_1_Species['Species'].isin(["['species_1' 'species_2']", "['species_2' 'species_1']"])]

    # Get Tag_1 with more than one Species in a dataframe
    Tag_1_more_than_1_Species =  df_Tag_1_Species.loc[df_Tag_1_Species['Species'].isin(["['species_1' 'species_2']", "['species_2' 'species_1']"])]
    Tag_1_more_than_1_Species = Tag_1_more_than_1_Species.Tag_1.tolist()
    return Tag_1_more_than_1_Species
```
